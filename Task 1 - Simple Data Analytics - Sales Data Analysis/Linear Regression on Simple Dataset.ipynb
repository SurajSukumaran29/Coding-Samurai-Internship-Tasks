#Linear Regression : Predicting Sales based on Advertising Spend

#Importing the Dataset

```python
import pandas as pd
df=pd.read_csv('/content/Advertising Budget and Sales.csv')
df
```

#Checking for Null Values

```python
df.isnull().sum()
```

#Checking the Datatypes

```python
df.dtypes
```

#Splitting to X features and y target variables

```python
X=df.iloc[:,0:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

#Splitting to Train and Test Set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

#Building the Model

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

#Predicting Using the Model

```python
y_pred=model.predict(X_test)
y_pred
```

#Calculating the Mean Squared Error and R2_Score

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print("Mean Squared Error =",mse)
print("r2_score =",r2)
```

#Plotting the Linear Regression:Actual Sales vs Predicted Sales

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred,color="blue",label="Predicted line")
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()],color='red',label="Actual Line")
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.legend()
plt.show()
```
