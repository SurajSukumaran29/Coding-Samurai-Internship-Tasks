```python
import pandas as pd
df=pd.read_csv('/content/twitter_validation.csv')
df
```

```python
df.columns = ['id', 'source', 'target', 'tweet']
df
```

#Plot the count of Positive,Negative and Neutral tweets

```python
import seaborn as sns
sns.countplot(x='target',data=df)
```

#Take only required columns

```python
df=df.loc[df['target']!='Irrelevant']
df
```

```python
df.reset_index(drop=True,inplace=True)
df
```

#Drop unwanted columns

```python
df.drop(['id','source'],axis=1,inplace=True)
df
```

#Map the target categorical values to numerical values

```python
#positive=1,negative=-1,neutral=0
df['target']=df['target'].map({'Positive':1,'Negative':-1,'Neutral':0})
df
```

#Text-Preprocessing

```python
tweets=df.tweet
tweets
```

#Remove unwanted characters

```python
tweets=tweets.str.replace('[^a-zA-Z0-9 ]',"",regex=True)
tweets
```

#Convert words to the root words using Stemming

```python
from nltk.stem import SnowballStemmer
stemmer=SnowballStemmer('english')
```

#Tokenize the text

```python
from nltk import TweetTokenizer
tk=TweetTokenizer()
tweets=tweets.apply(lambda x:[stemmer.stem(i.lower())for i in tk.tokenize(x)]).apply(lambda token:" ".join(token))
tweets
```

#Remove Stopwords

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
sw=stopwords.words('english')
print(sw)
```

```python
tweets=tweets.apply(lambda x:[i for i in tk.tokenize(x) if i not in sw]).apply(lambda token:" ".join(token))
tweets
```

#Vectorize the Preprocessed Texts

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
data=vectorizer.fit_transform(tweets)
print(data)
```

#Split to features and targets

```python
X=data.toarray()
X
```

```python
y=df['target']
y
```

#Train-Test Splitting

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

#Build the model and Predict

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
knn=KNeighborsClassifier(n_neighbors=5)
rf=RandomForestClassifier()
models=[knn,rf]
for model in models:
   print(model)
   model.fit(X_train,y_train)
   y_pred=model.predict(X_test)
   print(classification_report(y_test,y_pred))
   y_new=model.predict(vectorizer.transform(['csgo matchmak full closet hack truli aw game']))
   print(y_new)
   if y_new==-1:
    print("Negative")
   elif y_new==0:
    print("Neutral")
   else:
     print("Postive")
```
